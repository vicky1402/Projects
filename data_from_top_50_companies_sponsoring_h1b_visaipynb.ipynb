```python
from bs4 import BeautifulSoup
import requests
```

```python
url="https://www.myvisajobs.com/reports/h1b/"
```

```python
page=requests.get(url)
soup=BeautifulSoup(page.content,'html.parser')
print(soup.prettify() )
```

```python
table=soup.find('table')
rows=table.find_all(class_='headrow')
for row in rows:
  cols=row.find_all('td')
  cols=[x.text.strip() for x in cols]
  print(cols)
```

```python
import pandas as pd
df=pd.DataFrame(columns=cols)
```

```python
table_columns = len(table.find('tr').find_all('th'))

```

```python
# Loop through the rows and append data to the DataFrame
for row in rows[1:]:
    individual_data = row.find_all('td')
    individual_data_row = [data.text.strip() for data in individual_data]

    # Check if the number of elements in the individual_data_row matches the number of columns
    if len(individual_data_row) != table_columns:
        # Handle the case where the number of elements doesn't match
        # For example, you could skip the row or raise an exception
        continue

    # Append the data to the DataFrame
    df.loc[len(df)] = individual_data_row
```

```python
df.to_csv(r'/content/drive/MyDrive/Untitled folder/h1b.csv')
```
